```python
# Here we import all packages we need.

# This is a package for handling and manipulating data.
import pandas as pd

# This is the standard package for plotting.
from matplotlib import pyplot as plt 

# Seaborn is build on matplotlib.
# It is more comfortable to use than pure matplotlib.
import seaborn as sns

# Numpy helps as handling arrays of numbers
import numpy as np

# This computes the normal distribution's PDF
def normpdf(x, mean, sd):
    var = float(sd)**2
    denom = (2*np.pi*var)**.5
    num = np.exp(-((x)-(mean))**2/(2*var))
    return num/denom
```

## Logarithmic Scales

```python
# We load the planets dataset
planets = sns.load_dataset("planets")
planets
```

```python
# Let's make a scaterplot
sns.scatterplot(data=planets,x='mass',y='orbital_period')

# You can show multiple plots in one cell by calling this inbetween plots
plt.show()

# The planets are very cramped into the lower left corner
# We will try logarithmic scales
sns.scatterplot(data=planets,x='mass',y='orbital_period')
plt.yscale('log')

plt.show()
# logarithmic scales on both axis
sns.scatterplot(data=planets,x='mass',y='orbital_period')
plt.yscale('log')
plt.xscale('log')
```

## Modifying the Appearance of a Plot

```python
plt.show()
# logarithmic scales on both axis
# lets try to plot the detection method as well
sns.scatterplot(data=planets,x='mass',y='orbital_period', hue = 'year')
plt.yscale('log')
plt.xscale('log')
plt.show()

# We can better distinguish the points when we make them transparent.
sns.scatterplot(data=planets,x='mass',y='orbital_period', hue = 'year', alpha = 0.5)
plt.yscale('log')
plt.xscale('log')
plt.show()

# Let's get rid of the legend
sns.scatterplot(data=planets,x='mass',y='orbital_period', hue = 'year', alpha = 0.5, legend = False)
plt.yscale('log')
plt.xscale('log')
plt.show()
```

## Lineplots
You can make lineplots like this:

```python
# We load the flights dataset
flights = sns.load_dataset("flights")
flights
```

```python
# Let's average over all months of each year
flights= flights.groupby('year').mean()
flights
```

```python
sns.lineplot(data=flights, x='year', y='passengers')
```

## Saving a Figure
We can save a figure using the [savefig](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html) command.
This is useful, when we want to modify our figure later on using e.g. a vector graphics program.

```python
sns.lineplot(data=flights, x='year', y='passengers')
#plt.show()
# We can save it as png image
# The dpi parameter allows us to increase the resolution of the image 
plt.savefig('flights.png', dpi=300)

# We can also save it as pdf
plt.savefig('flights.pdf')

# We can also save it as svg
plt.savefig('flights.svg')
```
